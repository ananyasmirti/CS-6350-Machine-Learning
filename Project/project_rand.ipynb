```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
column = ['Age','Workclass','fnlgt','education ','education-num','marital-status','occupation','relationship','race','sex','capital-gain'
          ,'capital-loss','hours-per-week','native-country','class']

train =  pd.read_csv('./data/train_final.csv', names= column)
train=train.drop(index=0)
train.head(5)
```

```python
train.dtypes
```

```python
train.isnull().count()
```

```python
train.shape
```

```python
train['native-country'] = train['native-country'].replace(' ?' , np.nan )
train['Age'] = train['Age'].replace(' ?' , np.nan )
train['Workclass'] = train['Workclass'].replace(' ?' , np.nan )
train['fnlgt'] = train['fnlgt'].replace(' ?' , np.nan )
train['education-num'] = train['education-num'].replace(' ?' , np.nan )
train['marital-status'] = train['marital-status'].replace(' ?' , np.nan )
train['occupation'] = train['occupation'].replace(' ?' , np.nan )
train['relationship'] = train['relationship'].replace(' ?' , np.nan )
train['race'] = train['race'].replace(' ?' , np.nan )
train['sex'] = train['sex'].replace(' ?' , np.nan )
train['capital-gain'] = train['capital-gain'].replace(' ?' , np.nan )
train['capital-loss'] = train['capital-loss'].replace(' ?' , np.nan )
train['hours-per-week'] = train['hours-per-week'].replace(' ?' , np.nan )
```

```python
train.isna().sum()
```

```python
train['Workclass'] = train['Workclass'].fillna(train['Workclass'].mode()[0])
train['occupation'] = train['occupation'].fillna(train['occupation'].mode()[0])
train['native-country'] = train['native-country'].fillna(train['native-country'].mode()[0])
```

```python
train.describe()
```

```python
train.info()
```

```python
train.duplicated().sum()
```

```python
Data = train.drop_duplicates(keep= 'first')
Data.shape
```

```python
feat = list(train.select_dtypes(include=['object']).columns)
print(feat)
train
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for f in feat:
    train[f] = le.fit_transform(train[f].astype(str))
train
```

```python
X = train.drop(columns = ['class'])
y = train['class'].values
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 500)
model = classifier.fit(X, y)
```

```python
column = ['Age','Workclass','fnlgt','education ','education-num','marital-status','occupation','relationship','race','sex','capital-gain'
          ,'capital-loss','hours-per-week','native-country']

test =  pd.read_csv('./data/test_final.csv', names= column)
test = test.iloc[1: , :]
test.head(5)
```
